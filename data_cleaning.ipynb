```
#change - to 0
columns = ['date', 'cases', 'discharged', 'death', 'icu', 'pui', 'close-contact',
       'tabligh', 'surveillance', 'hadr', 'import', 'perlis', 'kedah',
       'pulau-pinang', 'perak', 'selangor', 'negeri-sembilan', 'melaka',
       'johor', 'pahang', 'terengganu', 'kelantan', 'sabah', 'sarawak',
       'wp-kuala-lumpur', 'wp-putrajaya', 'wp-labuan']

for i in columns:
  df[i] = df.apply(lambda x: 0 if x[i] == '-'  else x[i], axis=1)
```

```
#change NaN to previous value
df = df.fillna(method ='pad')
```

```
#change initial NaN with 0
df = df.fillna(0)
```
